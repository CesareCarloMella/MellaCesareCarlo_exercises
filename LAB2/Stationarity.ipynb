```python
import numpy as np
import matplotlib.pyplot as plt


from scipy.optimize import curve_fit

Npassi, Nsimul, Lpasso, Nblocchi = np.loadtxt("stationarity/Parameters.dat")

#x = np.linspace(0,100,100,endpoint=True)
print("Numero di blocchi =",Nblocchi)
fig=plt.figure(figsize=(12,8))
x, f , error=  np.loadtxt("stationarity/Risultati/LatticeData0.300000.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f, yerr=error, label="0.3")
plt.axhline(y=1./np.sqrt(1-0.3*0.3),color='r')
x, f , error=  np.loadtxt("stationarity/Risultati/LatticeData0.500000.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f, yerr=error, label="0.5")
plt.axhline(y=1./np.sqrt(1-0.5*0.5), color='r')
x, f , error=  np.loadtxt("stationarity/Risultati/LatticeData0.700000.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f, yerr=error, label="0.7")
plt.axhline(y=1./np.sqrt(1-0.7*0.7),color='r')
x, f , error=  np.loadtxt("stationarity/Risultati/LatticeData0.900000.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
#plt.errorbar(x,f, yerr=error, label="0.9")
#plt.axhline(y=1./np.sqrt(1-0.9*0.9),color='r')
x, f , error=  np.loadtxt("stationarity/Risultati/LatticeData0.990000.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
#plt.errorbar(x,f, yerr=error, label="0.99")
#plt.axhline(y=1./np.sqrt(1-0.99*0.99),color='r')
x,f, error= np.loadtxt("stationarity/Risultati/LatticeData1.000000.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
#plt.errorbar(x,f, yerr=error, label="1.00")

plt.legend()
plt.title('LatticeRandomWalk')
plt.xlabel('#passi')
plt.ylabel('distanza')
plt.grid(True)

print("Lunghezza del passo =", Lpasso)
print("Numero di passi =", Npassi)
print("Numero di simulazioni su cui medio =",Nsimul)
print("")


plt.show()
```
